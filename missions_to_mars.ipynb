```python
# Dependencies
from bs4 import BeautifulSoup as bs
from splinter import Browser
import requests
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

## NASA Mars News

```python
# Scrape the url "redplantscience.com"
url = 'https://redplanetscience.com/'
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(url)
```

```python
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
html = browser.html
```

```python
soup = bs(html, "html.parser")  
```

```python
# Used print(soup.prettify()) to observe html
print(soup.prettify())
```

```python
# Extract Title and Contents of latest article from html
news_title = soup.find(class_= 'content_title').text.strip()
news_p = soup.find(class_='article_teaser_body').text.strip()

# Print Results
print(f'News Title: "{news_title}"')
print(f'News Contents: "{news_p}"')
```

## JPL Mars Space Images - Featured Image

```python
url_2 = 'https://spaceimages-mars.com/'
```

```python
# Initiate Splinter and open browser at desired url
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(url_2)
```

```python
# Parse the webpage's html
html = browser.html
soup = bs(html, 'html.parser')
```

```python
# Identify the footer of the html and the elements within it
image_url_1 = soup.find('img', class_='headerimage fade-in')
image_url = image_url_1.get('src')

featured_image_url = 'https://spaceimages-mars.com/' + image_url
print(featured_image_url)
```

## Mars Hemispheres

```python
# Scrape the url 
url_4 = 'https://marshemispheres.com/'
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(url_4)
```

```python
html = browser.html
```

```python
soup = bs(html, "html.parser")  
print(soup.prettify())
```

```python
hemisphere_image_urls = []

links = browser.find_by_css('a.product-item img')

for i in range(len(links)):
    hemisphere = {}
    
    links[i].click()
    
    sample_img = browser.links.find_by_text('Sample').first
    
    hemisphere['img_url'] = sample_img['href']
    
    hemisphere['title'] = browser.find_by_css('h2.title').text
    
    hemisphere_image_urls.append(hemisphere)
    
    browser.back()
```
